```python
PATH = ""
```

```python
import pandas as pd
df = pd.read_csv(PATH+'hasil_preprocessing.csv', delimiter=',')
df
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df['sentimen'] = df['sentimen'].map({-1: 0, 0: 1, 1: 2})
df
```

```python
# Pastikan hanya ambil kolom 'lemmatization' dan 'sentimen'
df = df[['stemming', 'sentimen']]
df = df.rename(columns={'stemming': 'text', 'sentimen': 'label'})
df
```

```python
MAX_LEN = 512
TRAIN_BATCH_SIZE = 4
VALID_BATCH_SIZE = 2
EPOCHS = 10
LEARNING_RATE = 1e-05
N_CLASSES = 3
```

```python
from datasets import Dataset
```

```python
dataset = Dataset.from_pandas(df)
dataset
```

```python
dataset = dataset.train_test_split(test_size=0.2, seed=42)

train_dataset = dataset['train']
val_dataset   = dataset['test']
```

```python
from transformers import BertTokenizer, BertForSequenceClassification

model_name = "cahya/distilbert-base-indonesian"
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForSequenceClassification.from_pretrained(model_name, num_labels=3)  # num_labels sesuai jumlah kelas
```

```python
def tokenize(batch):
    return tokenizer(batch['text'], max_length=MAX_LEN, truncation=True)

train_dataset = train_dataset.map(tokenize, batched=True)
val_dataset = val_dataset.map(tokenize, batched=True)
```

```python
class_weights = (1 - (df["label"].value_counts().sort_index() / len(df))).values
class_weights
```

```python
import torch

class_weights = torch.from_numpy(class_weights).float().to("cuda")
class_weights
```

```python
val_dataset = val_dataset.rename_column("label", "labels")
train_dataset = train_dataset.rename_column("label", "labels")

val_dataset, train_dataset
```

```python
from transformers import Trainer, TrainingArguments
from torch import nn
```

```python
from sklearn.metrics import f1_score, accuracy_score

def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions
    accuracy = accuracy_score(labels, preds)
    return { "acc : " : accuracy}
```

```python
class WeightedLossTrainer(Trainer):
    def compute_loss(self, model, inputs, return_outputs=False, **kwargs):
        outputs = model(**inputs)
        logits = outputs.get("logits")
        labels = inputs.get("labels")
        loss_func = nn.CrossEntropyLoss(weight=class_weights)
        loss = loss_func(logits, labels)
        return (loss, outputs) if return_outputs else loss         
```

```python
import os
os.environ["WANDB_DISABLED"] = "true"

training_args = TrainingArguments(
    eval_strategy="epoch",              # Evaluasi setiap epoch
    save_strategy="epoch",                    # Simpan model setiap epoch
    load_best_model_at_end=True,              # Simpan model terbaik (berdasarkan metrik)
    metric_for_best_model="eval_loss",        # Gunakan val_loss sebagai acuan
    greater_is_better=False,                  # Lebih kecil = lebih baik (karena loss)
    save_total_limit=1,                       # Simpan hanya model terbaik
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=4,
    learning_rate=2e-5,
    weight_decay=0.01,
    lr_scheduler_type="cosine",
    max_grad_norm=1.0,
    logging_dir='./logs',
    logging_strategy="epoch",
    fp16=True,
    output_dir="./my_indodistilbert_sentimen",
    overwrite_output_dir=True,
    
)


# trainer = Trainer(
#     model=model,
#     args=training_args,
#     train_dataset=train_dataset,
#     eval_dataset=val_dataset,
#     tokenizer=tokenizer,
#     compute_metrics=compute_metrics,
# )

trainer = WeightedLossTrainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    tokenizer=tokenizer,
)

```

```python
trainer.train()
```

```python
import pandas as pd
import torch

tweet_post_df = pd.read_csv("./backend_api/data/tweet_post_4.csv", delimiter=",")
tweet_post_input_sample_list = tweet_post_df["Tweet"].sample(25).tolist()
    
tokenized_input = tokenizer(tweet_post_input_sample_list, return_tensors="pt", truncation=True, padding=True)
tokenized_input
```

```python
model.eval()

# Forward pass
with torch.no_grad():
    outputs = model(**tokenized_input)

    # The logits (raw model output)
    logits = outputs.logits

    predicted_class = torch.argmax(logits, dim=-1)

# Decode label (optional)
id2label = model.config.id2label

for text, pred in zip(tweet_post_input_sample_list, predicted_class):
    label = id2label.get(pred.item(), pred.item())
    print(f"Text: {text}\nPredicted Label: {label}\n")
```
